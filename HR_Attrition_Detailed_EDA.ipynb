```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset (from Google Drive)
df = pd.read_csv("/content/drive/MyDrive/HR-Employee-Attrition.csv")

# Map 'Attrition' to 0 and 1
df['Attrition'] = df['Attrition'].map({'No': 0, 'Yes': 1})

# Analyze attrition by Department
attrition_by_dept = df.groupby('Department')['Attrition'].mean().reset_index()
print("\nAttrition Rate by Department:")
print(attrition_by_dept)

# Visualize attrition by Department
plt.figure(figsize=(8, 6))
sns.barplot(x='Department', y='Attrition', data=attrition_by_dept)
plt.title('Attrition Rate by Department')
plt.ylabel('Attrition Rate')
plt.xlabel('Department')
plt.show()
```

```python
# Create salary bands
bins = [0, 3000, 6000, 9000, 12000, df['MonthlyIncome'].max()]
labels = ['0-3000', '3001-6000', '6001-9000', '9001-12000', '12000+']
df['SalaryBand'] = pd.cut(df['MonthlyIncome'], bins=bins, labels=labels, right=False)

# Analyze attrition by Salary Band
attrition_by_salary = df.groupby('SalaryBand')['Attrition'].mean().reset_index()
print("\nAttrition Rate by Salary Band:")
print(attrition_by_salary)

# Visualize attrition by Salary Band
plt.figure(figsize=(10, 6))
sns.barplot(x='SalaryBand', y='Attrition', data=attrition_by_salary)
plt.title('Attrition Rate by Salary Band')
plt.ylabel('Attrition Rate')
plt.xlabel('Salary Band')
plt.show()
```

```python
# Analyze attrition by Years Since Last Promotion
attrition_by_promotion = df.groupby('YearsSinceLastPromotion')['Attrition'].mean().reset_index()
print("\nAttrition Rate by Years Since Last Promotion:")
print(attrition_by_promotion)

# Visualize attrition by Years Since Last Promotion
plt.figure(figsize=(12, 6))
sns.barplot(x='YearsSinceLastPromotion', y='Attrition', data=attrition_by_promotion)
plt.title('Attrition Rate by Years Since Last Promotion')
plt.ylabel('Attrition Rate')
plt.xlabel('Years Since Last Promotion')
plt.show()
```
